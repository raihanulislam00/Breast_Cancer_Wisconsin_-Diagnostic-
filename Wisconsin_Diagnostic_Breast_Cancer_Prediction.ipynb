# Feature Importance
The importance of a feature is determined by how much the feature decreases the weighted impurity (e.g., Gini impurity or entropy) across all the trees in the forest. We can use the feature importance later if we need to reduce any features.

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(df.drop('target', axis=1), df['target'])

importances = model.feature_importances_
indices = np.argsort(importances)

plt.figure(figsize=(12, 8))
plt.title('Feature Importances')
plt.barh(range(len(importances)), importances[indices], align='center')
plt.yticks(range(len(importances)), [cancer.feature_names[i] for i in indices])
plt.xlabel('Feature Importance')
plt.show()
```

# Spliting The dataset into training and testing set
The splitting is done with 80-20 split. There are no validation set.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x=df.drop('target',axis=1)
y=df['target']
```

```python
df['target'].value_counts() /len(df)*100
```

```python
scaler=StandardScaler()
x=scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

# Model trainning
Now we will train the model using some most used classification models. the steps are:
1. In data visualization we have seen the dataset is positively skewed.It may impact models like SVM and KNN Stararizing the features using StandardScaler() function which uses the z score normalization.
2. the initilial classifier is the LogisticRegression
3. The param_grid defines all the classifiers used in the program with their parameter. The param_grid is a great way to use hyperparameter tuning and finding the best parameters for the model and the dataset.
4. Finally in grid_search we are using 10 k-fold cross validation with accuracy metric for the performance.
5. Then we trained the training data to find the best model.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
```

```python
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', LogisticRegression())
])
```

```python
param_grid = [
    {
        'classifier': [LogisticRegression()],
        'classifier__C': [0.01, 0.1, 1, 10, 100],
        'classifier__solver': ['liblinear', 'saga']
    },
    {
        'classifier': [RandomForestClassifier()],
        'classifier__n_estimators': [10, 50, 100, 200],
        'classifier__max_depth': [None, 10, 20, 30]
    },
    {
        'classifier': [SVC()],
        'classifier__C': [0.01, 0.1, 1, 10, 100],
        'classifier__kernel': ['linear', 'rbf']
    },
    {
        'classifier': [KNeighborsClassifier()],
        'classifier__n_neighbors': [3, 5, 7, 9],
        'classifier__weights': ['uniform', 'distance']
    },
        {
        'classifier': [GradientBoostingClassifier()],
        'classifier__n_estimators': [50, 100, 200],
        'classifier__learning_rate': [0.01, 0.1, 0.2],
        'classifier__max_depth': [3, 5, 7]
    },
    {
        'classifier': [XGBClassifier()],
        'classifier__n_estimators': [100, 200, 300],
        'classifier__learning_rate:': [0.01, 0.1, 0.2],
        'classifier__max_depth': [3, 5, 7]
    }

]
```

```python

grid_search = GridSearchCV(pipeline, param_grid, cv=10, scoring='accuracy', n_jobs=-1, verbose=1)
grid_search.fit(x_train, y_train)
```

```python
print("Best Parameters:", grid_search.best_params_)
print("Best Cross-Validation Accuracy:", grid_search.best_score_)
```

```python

best_model = grid_search.best_estimator_
test_score = best_model.score(x_test, y_test)
print("Test Set Accuracy:", test_score)
```

The best model is the Logistic Regression with low regulazation and liblinear cost optimization function.

Now Let's see the best results from all the models with the best parameters

```python
results_df = pd.DataFrame(grid_search.cv_results_)

results_df['param_classifier'] = results_df['param_classifier'].apply(lambda x: x.__class__.__name__)

best_results = results_df.loc[results_df.groupby('param_classifier')['mean_test_score'].idxmax()]

test_scores_with_params = {}

for _, row in best_results.iterrows():
    clf_name = row['param_classifier']
    best_params = row['params']
    pipeline.set_params(**best_params)
    pipeline.fit(x_train, y_train)
    test_score = pipeline.score(x_test, y_test)
    test_scores_with_params[clf_name] = {
        'Test Accuracy': test_score,
        'Best Parameters': best_params
    }
for clf_name, result in test_scores_with_params.items():
    print(f"Best Test Set Accuracy for {clf_name}: {result['Test Accuracy']:.4f}")
    print(f"Best Parameters: {result['Best Parameters']}")
    print("-" * 50)
```

Let's plot a classfication report including precision, recall, f-1 score and support across 0,1,average accuracy, micro accuracy and weighted accuracy.
Also let's find out the confusion matric and plot them

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
evaluation_metrics = {}

for _, row in best_results.iterrows():
    clf_name = row['param_classifier']
    best_params = row['params']

    pipeline.set_params(**best_params)

    pipeline.fit(x_train, y_train)

    y_pred = pipeline.predict(x_test)

    test_accuracy = pipeline.score(x_test, y_test)
    report = classification_report(y_test, y_pred, output_dict=True)
    conf_matrix = confusion_matrix(y_test, y_pred)

    evaluation_metrics[clf_name] = {
        'Test Accuracy': test_accuracy,
        'Classification Report': report,
        'Confusion Matrix': conf_matrix,
        'Best Parameters': best_params
    }

for clf_name, metrics in evaluation_metrics.items():
    print(f"Best Test Set Accuracy for {clf_name}: {metrics['Test Accuracy']:.4f}")
    print(f"Best Parameters: {metrics['Best Parameters']}")
    print("\nClassification Report:")
    print(pd.DataFrame(metrics['Classification Report']).transpose())
    print("\nConfusion Matrix:")
    print(metrics['Confusion Matrix'])
    print("-" * 50)
```

# overall performance measure with accuracy_score, precision_score, recall_score, matthews_correvoef, f1_score

```python

from sklearn.metrics import accuracy_score, precision_score, recall_score, matthews_corrcoef, f1_score

performance_metrics = {}

for clf_name, metrics in evaluation_metrics.items():
    y_pred = pipeline.predict(x_test)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    mcc = matthews_corrcoef(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    tn, fp, _, _ = metrics['Confusion Matrix'].ravel()
    specificity = tn / (tn + fp)

    performance_metrics[clf_name] = {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'MCC': mcc,
        'Specificity': specificity,
        'F1-score': f1
    }

for clf_name, metrics in performance_metrics.items():
    print(f"Performance Metrics for {clf_name}:")
    for metric_name, value in metrics.items():
        print(f"{metric_name}: {value:.4f}")
    print("-" * 50)
```

```python
metrics_data = {}
for clf_name, metrics in evaluation_metrics.items():
    metrics_data[clf_name] = {
        'Accuracy': metrics['Test Accuracy'],
        'Precision': metrics['Classification Report']['weighted avg']['precision'],
        'Recall': metrics['Classification Report']['weighted avg']['recall'],
        'F1-Score': metrics['Classification Report']['weighted avg']['f1-score'],
        'Specificity': metrics['Classification Report']['0']['recall'],
        'MCC': matthews_corrcoef(y_test, pipeline.predict(x_test))
    }

metrics_df = pd.DataFrame(metrics_data).transpose()

fig, ax = plt.subplots(figsize=(10, 6))
metrics_df.plot(kind='bar', ax=ax)
ax.set_title('Performance Metrics for Different Models')
ax.set_ylabel('Score')
ax.set_xticklabels(metrics_df.index, rotation=45)
ax.legend(loc='lower right')
ax.grid(True)
plt.tight_layout()
plt.show()
```

```python
num_classifiers = len(evaluation_metrics)
cols = 2
rows = (num_classifiers + 1) // cols
fig, axes = plt.subplots(rows, cols, figsize=(15, 5 * rows))
axes = axes.flatten()

for i, (clf_name, metrics) in enumerate(evaluation_metrics.items()):
    conf_matrix = metrics['Confusion Matrix']

    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='coolwarm',
                xticklabels=['Class 0', 'Class 1'],
                yticklabels=['Class 0', 'Class 1'],
                ax=axes[i])
    axes[i].set_xlabel('Predicted Label')
    axes[i].set_ylabel('True Label')
    axes[i].set_title(f'Confusion Matrix for {clf_name}')

for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```

# Accuracy and error plotting

```python
best_accuracies = {clf_name: metrics['Test Accuracy'] for clf_name, metrics in evaluation_metrics.items()}
plt.figure(figsize=(10, 6))
plt.bar(best_accuracies.keys(), best_accuracies.values())
plt.xlabel('Classifier')
plt.ylabel('Best Test Accuracy')
plt.title('Best Test Accuracies for Different Classifiers')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python

import matplotlib.pyplot as plt
train_errors = {}
test_errors = {}

for clf_name, metrics in evaluation_metrics.items():
    best_params = metrics['Best Parameters']
    pipeline.set_params(**best_params)

    pipeline.fit(x_train, y_train)

    train_pred = pipeline.predict(x_train)
    test_pred = pipeline.predict(x_test)

    train_errors[clf_name] = 1 - accuracy_score(y_train, train_pred)
    test_errors[clf_name] = 1 - accuracy_score(y_test, test_pred)

plt.figure(figsize=(10, 6))
plt.bar(train_errors.keys(), train_errors.values(), color='blue', label='Training Error')
plt.bar(test_errors.keys(), test_errors.values(), color='red', label='Testing Error', bottom=list(train_errors.values()))

plt.xlabel('Classifier')
plt.ylabel('Error Rate')
plt.title('Training and Testing Errors for Different Classifiers')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```
